```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('D:\DOCUMENTS\car_crashes.csv')
```

```python
data.head()
```

```python
data.info()
```

univariate

```python
plt.figure(figsize=(12, 6))
plt.subplot(2, 3, 1)
sns.histplot(data['total'], kde=True)
plt.title('Distribution of Total')
plt.subplot(2, 3, 2)
sns.histplot(data['speeding'], kde=True)
plt.title('Distribution of Speeding')
plt.subplot(2, 3, 3)
sns.histplot(data['alcohol'], kde=True)
plt.title('Distribution of Alcohol')
plt.subplot(2, 3, 4)
sns.histplot(data['ins_premium'], kde=True)
plt.title('Distribution of Insurance Premium')
plt.subplot(2, 3, 5)
sns.histplot(data['ins_losses'], kde=True)
plt.title('Distribution of Insurance Losses')
plt.tight_layout()
plt.show()
```

bivariate


```python
plt.figure(figsize=(12, 6))
plt.subplot(2, 3, 1)
sns.scatterplot(x='total', y='speeding', data=data)
plt.title('Total vs. Speeding')
plt.subplot(2, 3, 2)
sns.scatterplot(x='alcohol', y='speeding', data=data)
plt.title('Alcohol vs. Speeding')
plt.subplot(2, 3, 3)
sns.scatterplot(x='ins_premium', y='ins_losses', data=data)
plt.title('Insurance Premium vs. Insurance Losses')
plt.subplot(2, 3, 4)
sns.scatterplot(x='not_distracted', y='no_previous', data=data)
plt.title('Not Distracted vs. No Previous')
plt.tight_layout()
plt.show()
```

multivariate

```python
sns.pairplot(data)
plt.show()
```
